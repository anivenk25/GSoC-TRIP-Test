```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data
data = pd.read_csv('/kaggle/input/trip-test/Experimenter_9110002_53.dat', delimiter='\s+', header=0)

# Convert numeric columns to numeric type, coercing errors to NaN
data = data.apply(pd.to_numeric, errors='coerce')

# Drop rows with missing values in relevant columns
data.dropna(subset=['HeadwayDistance', 'Velocity', 'LeadVehPosX', 'LeadVehPosY'], inplace=True)

# Calculate relative velocity based on lead vehicle position
lead_vehicle_speed = ((data['LeadVehPosX'] ** 2 + data['LeadVehPosY'] ** 2) ** 0.5).diff()  # Assuming consecutive rows represent consecutive time steps
relative_velocity = lead_vehicle_speed.diff()

# Select features for anomaly detection
anomaly_features = ['HeadwayDistance', 'FollowCarBrakingStatus']

# Train the Isolation Forest model for anomaly detection
isolation_forest = IsolationForest(contamination=0.05)  # Adjust contamination based on expected proportion of anomalies
isolation_forest.fit(data[anomaly_features])

# Predict outliers (anomalies)
anomaly_predictions = isolation_forest.predict(data[anomaly_features])

# Filter the dataset to include only non-anomalous instances
non_anomalous_data = data[anomaly_predictions == 1]

# Calculate Time to Collision (TTC) metric for safe following distance
non_anomalous_data['Time_to_Collision'] = non_anomalous_data['HeadwayDistance'] / abs(relative_velocity)

# Define a threshold for safe following distance (TTC) using the three-second rule
safe_distance_threshold = 3 * non_anomalous_data['Velocity']  # Three-second rule: Safe distance = 3 * velocity

# Identify instances where the following vehicle is too close to the lead vehicle
unsafe_following = non_anomalous_data[non_anomalous_data['Time_to_Collision'] < safe_distance_threshold]

# Print the number of unsafe following instances
print("Number of instances with unsafe following distance (using three-second rule):", len(unsafe_following))

# Plotting Time to Collision (TTC)
plt.figure(figsize=(10, 6))
plt.plot(non_anomalous_data.index, non_anomalous_data['Time_to_Collision'], color='blue', label='TTC')
plt.scatter(unsafe_following.index, unsafe_following['Time_to_Collision'], color='red', label='Unsafe Following')

plt.title('Time to Collision (TTC) vs. Data Index')
plt.xlabel('Data Index')
plt.ylabel('Time to Collision (TTC)')
plt.legend()
plt.grid(True)
plt.show()

# Extract and print the 'MediaTime' values for instances of unsafe following distance
unsafe_following_media_time = unsafe_following['MediaTime']
print("MediaTime for instances with unsafe following distance:")
print(unsafe_following_media_time)


```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Read the data
data = pd.read_csv('/kaggle/input/trip-test/Experimenter_9110002_53.dat', delimiter='\s+', header=0)

# Convert numeric columns to numeric type, coercing errors to NaN
data = data.apply(pd.to_numeric, errors='coerce')

# Drop rows with missing values in relevant columns
data.dropna(subset=['LonAccel', 'LatAccel', 'Throttle', 'Brake'], inplace=True)

# Select features for anomaly detection
anomaly_features = ['LonAccel', 'LatAccel', 'Throttle', 'Brake']

# Train the Isolation Forest model for anomaly detection
isolation_forest = IsolationForest(contamination=0.05)  # Adjust contamination based on expected proportion of anomalies
isolation_forest.fit(data[anomaly_features])

# Predict outliers (anomalies)
anomaly_predictions = isolation_forest.predict(data[anomaly_features])

# Filter the dataset to include only non-anomalous instances
non_anomalous_data = data[anomaly_predictions == 1]

# Plot anomalies over time for each metric
plt.figure(figsize=(12, 8))

# Plot LonAccel
plt.subplot(2, 2, 1)
plt.plot(data.index, data['LonAccel'], color='blue', label='LonAccel')
plt.scatter(data.index[anomaly_predictions == -1], data['LonAccel'][anomaly_predictions == -1], color='red', label='Anomaly')
plt.title('LonAccel Anomalies over Time')
plt.xlabel('Time')
plt.ylabel('LonAccel')
plt.legend()
plt.grid(True)

# Plot LatAccel
plt.subplot(2, 2, 2)
plt.plot(data.index, data['LatAccel'], color='green', label='LatAccel')
plt.scatter(data.index[anomaly_predictions == -1], data['LatAccel'][anomaly_predictions == -1], color='red', label='Anomaly')
plt.title('LatAccel Anomalies over Time')
plt.xlabel('Time')
plt.ylabel('LatAccel')
plt.legend()
plt.grid(True)

# Plot Throttle
plt.subplot(2, 2, 3)
plt.plot(data.index, data['Throttle'], color='orange', label='Throttle')
plt.scatter(data.index[anomaly_predictions == -1], data['Throttle'][anomaly_predictions == -1], color='red', label='Anomaly')
plt.title('Throttle Anomalies over Time')
plt.xlabel('Time')
plt.ylabel('Throttle')
plt.legend()
plt.grid(True)

# Plot Brake
plt.subplot(2, 2, 4)
plt.plot(data.index, data['Brake'], color='purple', label='Brake')
plt.scatter(data.index[anomaly_predictions == -1], data['Brake'][anomaly_predictions == -1], color='red', label='Anomaly')
plt.title('Brake Anomalies over Time')
plt.xlabel('Time')
plt.ylabel('Brake')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Calculate Aggressive Driving Score
aggressive_driving_score = ((data['Throttle'] - data['Brake']).abs() + data['LonAccel'].abs() + data['LatAccel'].abs()).mean()

# Calculate Following Distance Score
following_distance_score = data['HeadwayDistance'].mean()  # Assuming HeadwayDistance is available in the data

# Calculate Speeding Score
speeding_score = (data['Velocity'] > 40).mean()  # Define speed_limit_threshold

# Calculate Anomaly Detection Score
anomaly_detection_score = (anomaly_predictions == -1).mean()

# Combine scores (you can adjust weights based on importance)
weight_aggressive = 0.2
weight_following_distance = 0.3
weight_speeding = 0.3
weight_anomaly_detection = 0.2
driver_risk_score = (
    aggressive_driving_score * weight_aggressive +
    following_distance_score * weight_following_distance +
    speeding_score * weight_speeding +
    anomaly_detection_score * weight_anomaly_detection
)

print("Driver Risk Score:", driver_risk_score)

```
